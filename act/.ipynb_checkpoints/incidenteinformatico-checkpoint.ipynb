# Regresión Lineal: Coste de un incidente de seguridad

En este ejercicio se explican los fundamentos básicos de la regresión lineal aplicada a un caso de uso sencillo relacionado con la Ciberseguridad.

## Enunciado del ejercicio

El ejercicio consiste en predecir el coste de un incidente de seguridad en base al número de equipos que se han visto afectados. El conjunto de datos es generado de manera aleatoria.

### 0. Imports

```python
# Instalacion de librerías externas
!pip install pandas
!pip install numpy
!pip install matplotlib
!pip install scikit-learn
```

### 1. Generación del conjunto de datos

```python
import numpy as np

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

print("La longitud del conjunto de datos es:", len(X))
```

### 2. Visualización del conjunto de datos

```python
# tu respuesta aqui
import matplotlib.pyplot as plt
plt.plot(X, y, ".")
plt.title("Incidentes de seguridad")
plt.xlabel("Número de equipos")
plt.ylabel("Coste aprox")
plt.grid()
plt.show()
```

### 3. Modificación del conjunto de datos

```python
# Tu respuesta aqui
import pandas as pd
# Convertimos a DataFrame para ver mejor los datos
df = pd.DataFrame(np.hstack((X, y)), columns=["Equipos", "Coste"])
print(df.head(9))
```

```python
# Escalado del número de equipos afectados
df["Equipos"] = df["Equipos"] * 100
df["Equipos"] = df["Equipos"].astype("int")
# Escalado del coste
df["Coste"] = df["Coste"] * 1000
df["Coste"] = df["Coste"].astype("int")
df.head(10)
```

### 4. Construcción del modelo

```python
#Tu respuesta aqui
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(df["Equipos"].values.reshape(-1, 1), df["Coste"].values)

modelo.intercept_

modelo.coef_

X_min_max = np.array([[df["Equipos"].min()], [df["Equipos"].max()]])
y_train_pred = modelo.predict(X_min_max)

plt.plot(X_min_max, y_train_pred)
plt.grid()
plt.show()
```

### 5. Predicción de nuevos ejemplos

```python
#Tu respuesta aqui
X_nuevo = np.array([[52]])  # nuevos valores de entrada
y_pred = modelo.predict(X_nuevo)

# Mostrar predicciones
print("El coste sería:", int(y_pred[0]))
```

```python
# Mostrar la línea de regresión sobre los datos
plt.scatter(df["Equipos"], df["Coste"], color="green", label="Datos reales")
plt.plot(X_min_max, y_train_pred, color="red", label="Línea de regresión")
plt.plot(X_nuevo, y_pred, "X", color="black")
plt.title("Modelo de Regresión Lineal")
plt.xlabel("Número de equipos")
plt.ylabel("Coste aproximado")
plt.legend(loc="best")
plt.grid()
plt.show()
```
